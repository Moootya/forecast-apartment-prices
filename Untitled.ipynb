```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline

from attributes_adder import CombinedAttributesAdder

%matplotlib inline
```

```python
# Ограничиваем категории дохода путем деления медианного дохода на 1.5
data['income_cat'] = np.ceil(data['median_income'] / 1.5)
data['income_cat'].where(data['income_cat'] < 5, 5.0, inplace=True)


data['income_cat'].hist(bins=50, figsize=(20, 15))
```

```python
# Разбивка набора на тестовый и испытательный наборы по стратам
split = StratifiedShuffleSplit(n_splits=1, test_size=.2, random_state=42)


for train_index, test_index in split.split(data, data['income_cat']):
    strat_train_set = data.loc[train_index]
    strat_test_set = data.loc[test_index]
```

```python
for _set in (strat_train_set, strat_test_set):
    _set.drop('income_cat', axis=1, inplace=True)
```

```python
housing = strat_train_set.copy()
```

```python
# Общее значение комнат на округ — это бесполезная дичь, и надо бы получить кол-во комнат на дом/квартиру
housing['rooms'] = housing['total_rooms'] / housing['households']
```

```python
# Общее значение спален на округ — тоже сомнительно, надо получить количество спален для каждого дома
housing['bedrooms'] = housing['total_bedrooms'] / housing['total_rooms']
```

```python
# Количество людей на дом 
housing['population_per_household'] = housing['population'] / housing['households']
```

```python
# Замечаем сильную отрицательную корреляцию к количеству спален 
corr_matrix = housing.corr()
corr_matrix['median_house_value'].sort_values(ascending=False)
```
